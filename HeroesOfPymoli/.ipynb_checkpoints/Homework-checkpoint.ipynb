```python
# Percentage and Count of Male Players, Female Players, and Other/Non-Disclosed

gender = purchase_data[["SN", "Gender"]]

gender = gender.drop_duplicates()

counts = gender["Gender"].value_counts()

total_counts = [counts[0],counts[1],counts[2]]
percents = [round((counts[0]/player)*100,2),round((counts[1]/player)*100,2),round((counts[2]/player)*100,2)]

# Create a summary data frame 
gender_demo_df = pd.DataFrame(purchase_data["Gender"].value_counts())
gender_demo_df

# Add percentages to the summary data frame 
gender_demo = pd.DataFrame({ "Percentage of Players": percents,"Total Count": total_counts})
gender_demo.index = (["Male", "Female", "Other / Non-Disclosed"])
gender_demo
```

```python
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

gender = purchase_data.groupby(["Gender"])

purch_count = gender["SN"].count()

purch_price = gender["Price"].mean()

purch_value = gender["Price"].sum()

# Optional: give the displayed data cleaner formatting

duplicates = purchase_data.drop_duplicates(subset='SN', keep="first")
grouped_dup = duplicates.groupby(["Gender"])

purch_clean = (gender["Price"].sum() / grouped_dup["SN"].count())

# Create a summary data frame to hold the results

purch_df = pd.DataFrame({"Purchase Count": purch_count,
                              "Average Purchase Price": purch_price,
                              "Total Purchase Value": purch_value,
                              "Normalized Totals": purch_clean})

# Display the summary data frame

purch_df["Average Purchase Price"] = purch_df["Average Purchase Price"].map("${:.2f}".format)
purch_df["Total Purchase Value"] = purch_df["Total Purchase Value"].map("${:.2f}".format)
purch_df["Normalized Totals"] = purch_df["Normalized Totals"].map("${:.2f}".format)
purch_df = purch_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Normalized Totals"]]
purch_df
```
